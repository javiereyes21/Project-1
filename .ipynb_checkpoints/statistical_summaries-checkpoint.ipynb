```python
#dependencies
import pandas as pd
import requests
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cleaned_device_data = pd.read_csv("cleaned_device_data.csv")
cleaned_device_data
```

```python
# #Statistical summary (Mean, Median, Var, StdDev, SEM)UsageHoursPerDay

# # Calculate Mean 
usage_mean = cleaned_device_data['UsageHoursPerDay'].mean()

# # Calculate Median
usage_median = cleaned_device_data['UsageHoursPerDay'].median()

# # Calculate Variance
usage_variance = np.var(cleaned_device_data['UsageHoursPerDay'])

# # Calculate Standard Deviation
usage_std_dev = np.std(cleaned_device_data['UsageHoursPerDay'])

# # Calculate Standard Error of the Mean (SEM)
usage_sem = usage_std_dev / np.sqrt(len(cleaned_device_data))

print (usage_mean)
print (usage_median)
print(usage_variance)
print(usage_std_dev)
print(usage_sem)
```

```python
usage_stats = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Variance', 'Standard Deviation', 'SEM'],
    'Value': [usage_mean, usage_median, usage_variance, usage_std_dev, usage_sem]
})

plt.figure(figsize=(8, 4))
plt.bar(usage_stats['Statistic'], usage_stats['Value'], color='skyblue')
plt.xlabel('Statistics', fontweight='bold')
plt.ylabel('Values')
plt.title('Hours Per Day Usage', fontweight='bold')
plt.savefig('usage_stats', dpi=300, bbox_inches='tight')
plt.show()
```

```python
grouped_data = cleaned_device_data.groupby("DeviceType").agg(["mean", "median", "var", "std", "sem"])
grouped_data = grouped_data.reset_index()
```

```python
melted_data = grouped_data.melt(id_vars='DeviceType', var_name=['Metric', 'Stat'], value_name='Value')

plt.figure(figsize=(10, 8))
sns.barplot(x='DeviceType', y='Value', hue='Stat', data=melted_data,)

plt.title('Aggregated Statistics by DeviceType', fontweight='bold')
plt.xlabel('Device Type')
plt.ylabel('Value')
plt.legend(title='Statistic', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=5)
plt.tight_layout()

plt.show()
```
